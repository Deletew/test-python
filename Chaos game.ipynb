```python
import stddraw
import random
```

```python
n = 10000
```

```python
cx=[0.0,1.0,0.500]
cy=[0.0,0.0,0.866]

color=[stddraw.BLUE,stddraw.RED,stddraw.GREEN]
x,y=0,0
```

```python
for i in range(n):
    r=random.randrange(0,3)
    x=( x+cx[r] )/2.0
    y=( y+cy[r] )/2.0
    #stddraw.setPenColor(color[r])
    
    stddraw.point(x,y)
    
    stddraw.show(2)
    
stddraw.show()
```
