```python
import stddraw 
import random
```

```python
def TGR_one_Round(Stake,Goal): #Run once The Gambler's ruin
    long=0
    stddraw.setPenRadius(0.08)
    stddraw.setXscale(0,100)
    stddraw.setYscale(0,25)
    stddraw.line(0,Goal,500, Goal)
    stddraw.line(0,0,500, 0)
    cash=Stake
    cashb=Stake
    while cash>0 and cash<Goal :
        if random.randrange(0,2)==0:
            cash +=1
        else:
            cash -=1
        stddraw.line(long,cashb,long+1,cashb)
        stddraw.line(long+1,cashb,long+1, cash)
        long+=1
        cashb = cash
        stddraw.show(100)
    stddraw.show()

    
```

```python
def TGR(Stake):  #The Gambler's ruin
    cash=Stake
    while cash>0 and cash<Goal :
        if random.randrange(0,2)==0:
            cash +=1
        else:
            cash -=1
    return cash
    
```

```python
Stake=int(input()) #Start with this amount of money
Goal=int(input()) #win the race
Trials=int(input()) #Number of tests
```

## Possibility The Gambler's ruin Using estimates Monte Carlo method

```python
wins=0
bets=0

for t in range(Trials):
    if TGR(Stake)==Goal:
        wins+=1
        
print(str(100*wins /Trials))
TGR_one_Round(Stake,Goal)
```
